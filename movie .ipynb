```python
import pandas as pd

def categorize_col(df, col, labels):
    # Get the descriptive statistics for the column
    min_val = df[col].describe()['min']
    q1 = df[col].describe()['25%']
    median = df[col].describe()['50%']
    q3 = df[col].describe()['75%']
    max_val = df[col].describe()['max']
    
    # Define the edges based on the quartiles and min/max values
    edges = [min_val, q1, median, q3, max_val]
    
    # Apply pd.cut() for categorization
    df[col] = pd.cut(df[col], bins=edges, labels=labels, right=True, duplicates='drop')
    
    return df

# Example usage
labels = ['not_popular', 'below_avg', 'avg', 'popular']
df = categorize_col(df, 'Vote_Average', labels)

# Check unique values after categorization
print(df['Vote_Average'].unique())
```
